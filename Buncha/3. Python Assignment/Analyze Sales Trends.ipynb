```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sales_data.csv')
```

```python
df
```

```python
daily_sales = df.groupby('order_date')['revenue'].sum().reset_index()
```

```python
daily_sales
```

```python
# Plot sales trends
plt.figure(figsize=(10, 6))
plt.plot(daily_sales['order_date'], daily_sales['revenue'], marker='o')
plt.title('Daily Sales Trends')
plt.xlabel('Date')
plt.ylabel('Revenue')
#plt.grid(True)
plt.show()
```
